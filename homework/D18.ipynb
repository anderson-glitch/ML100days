```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#1. 畫出 cos 圖檔，並儲存
x = np.arange(0, 3 * np.pi, 0.1)
y = np.cos(x)
plt.plot(x, y)
plt.savefig("filename.png",dpi=300,format="png")
```

```python
#2. 給定散點圖顏色
#答案
n = 1024
x = np.random.normal(0,1,n)
y = np.random.normal(0,1,n)
color = np.arctan2(y,x)
plt.scatter(x, y, s = 75, c = color, alpha = 0.5)
plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)
plt.show()
```

```python
#練習結果
n = 1024
x = np.random.normal(0,1,n)
y = np.random.normal(0,1,n)
for i in range (0, 1024):
    if (x[i] < 0) & (y[i] > 0): plt.scatter(x[i], y[i], s = 75, color = 'yellow', alpha = 0.5)
    elif (x[i] > 0) & (y[i] > 0): plt.scatter(x[i], y[i], s = 75, color = 'green', alpha = 0.5)
    elif (x[i] > 0) & (y[i] < 0): plt.scatter(x[i], y[i], s = 75, color = 'blue', alpha = 0.5)
    elif (x[i] < 0) & (y[i] < 0): plt.scatter(x[i], y[i], s = 75, color = 'purple', alpha = 0.5)
plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)
plt.show()
```
